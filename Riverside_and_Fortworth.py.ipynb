```python
import os
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data_file = pd.read_csv("./FBI_UCR.csv")
data_file.head()
```

```python
data_file.dtypes
```

```python
fbi_count = data_file.loc[:,['name','variable','value','year','crime_type','place_name']]
fbi_count
```

```python
riverside_count = data_file.loc[data_file['name'] == 'Riverside, California']
pd.options.display.max_rows = 999
riverside_count["place_name"].fillna("Riverside, CA", inplace = True)
riverside_count
#file_path = os.path.join("Riverside.csv")
#riverside_count.to_csv(file_path, index=False)



```

```python
fortworth_count = data_file.loc[data_file['name'] == 'Fort Worth, Texas']
pd.options.display.max_rows = 999
fortworth_count["place_name"].fillna("Fort Worth, TX", inplace = True)
fortworth_count
#file_path = os.path.join("Riverside.csv")
#riverside_count.to_csv(file_path, index=False)
```

```python
RiversidecrimeRates = riverside_count.loc[riverside_count["variable"] == "rate"]
RiversidecrimeRates
```

```python
crimeTypes = RiversidecrimeRates["crime_type"].unique()
crimeTypes
```

```python
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018]

RiversideviolentCrimeRates = RiversidecrimeRates.loc[RiversidecrimeRates["crime_type"] == "Violent crime"]
RiversideviolentCrimeRates = RiversideviolentCrimeRates.drop_duplicates()
RiversideviolentCrimeRatesYrs = RiversideviolentCrimeRates.loc[RiversideviolentCrimeRates["year"].isin(years)]
RiversideviolentCrimeRatesYrs
```

```python
FortworthcrimeRates = fortworth_count.loc[fortworth_count["variable"] == "rate"]
FortworthcrimeRates
```

```python
crimeTypes = FortworthcrimeRates["crime_type"].unique()
crimeTypes
```

```python
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018]

FortworthviolentCrimeRates = FortworthcrimeRates.loc[FortworthcrimeRates["crime_type"] == "Violent crime"]
FortworthviolentCrimeRates = FortworthviolentCrimeRates.drop_duplicates()
FortworthviolentCrimeRatesYrs = FortworthviolentCrimeRates.loc[FortworthviolentCrimeRates["year"].isin(years)]
FortworthviolentCrimeRatesYrs
```

```python
plt.hist(FortworthviolentCrimeRates['year'])
plt.xlabel('year')
plt.ylabel('value')
plt.title("Fortworth Violent Crime")
plt.show()
```
