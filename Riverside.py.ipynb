```python
years = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]

RiversidecrimeRates = riverside_count.loc[riverside_count["variable"] == "rate"]

# creating violent crime only set from there:
# restrict this violent crime set to a period of years useful for comparison w/ weed-legal cities:
RiversidecrimeRatesYears = RiversidecrimeRates.loc[RiversidecrimeRates["year"].isin(years)]
RiversidecrimeRatesYears = RiversidecrimeRatesYears.drop_duplicates()

# remove extraneous columns
RiversidecrimeRatesYears = RiversidecrimeRatesYears.drop(columns=["key", "id", "place_name"])
# sort the years to print in ascending order
RiversidecrimeRatesYears = RiversidecrimeRatesYears.sort_values(['year'], ascending=[True])

Vcrime = ["Forcible rape", "Murder and nonnegligent manslaughter", "Aggravated assault", "Rape (revised definition)"]
RiversidecrimeRatesViolent = RiversidecrimeRatesYears.loc[RiversidecrimeRatesYears["crime_type"].isin(Vcrime)]

# Create seperate dataframes for each type of crime:

# create another category for the rape-specific crime types
rapeCategories = ["Forcible rape", "Rape (revised definition)"]
Riversiderape = RiversidecrimeRatesViolent.loc[RiversidecrimeRatesViolent["crime_type"].isin(rapeCategories)]
# set the index of each dataframe to the year the data was from
Riversiderape = Riversiderape.set_index(["year"])

Riversidemurder = RiversidecrimeRatesViolent.loc[RiversidecrimeRatesViolent["crime_type"] == "Murder and nonnegligent manslaughter"]
Riversidemurder = Riversidemurder.set_index(["year"])

Riversideassault = RiversidecrimeRatesViolent.loc[RiversidecrimeRatesViolent["crime_type"] == "Aggravated assault"]
Riversideassault = Riversideassault.set_index(["year"])

# Create a stacked bar plot where each crime type is in a distinct color:

# indices of each dataframe are the same size, can use arbitrary one for plot x-axis
Riverside_x_axis = Riversiderape.index
# each bar is a line here
p1 = plt.bar(Riverside_x_axis, Riversiderape["value"], color='black', alpha=0.5, align="center")
# the bottom arguments indicate where to start the second, third, etc bar in the set. From 2 on they become sums:
p2 = plt.bar(Riverside_x_axis, Riversidemurder["value"].values, color='#AD1457', bottom=Riversiderape["value"].values)
p3 = plt.bar(Riverside_x_axis, Riversideassault["value"].values, color='#4527A0', bottom=(Riversidemurder["value"] + Riversiderape["value"]).values)

tick_locations = [value for value in Riverside_x_axis]
plt.xticks(tick_locations, years)
plt.title("Riverside, California Violent Crime Rate: \n Comp Over Pre- and Post-Marijuana Legalization Years")
plt.ylabel("Crime Rate (per 100,000 People)")
plt.legend((p1[0], p2[0], p3[0]), ("Rape", "Murder & Non-negligent Manslaughter", "Aggravated Assault"))
RiversidecrimeRatesViolent
```

```python
# Non-violent crime categorization for NY starts here:
# create non-violent crimes list
nonVcrime = ["Motor vehicle theft", "Larceny", "Burglary", "Robbery", "Property crime"]

# restricting new DataFrame to non-violent crimes in above list:
RiversidecrimeRatesNonViolent = RiversidecrimeRates.loc[RiversidecrimeRates["crime_type"].isin(nonVcrime)]
RiversidecrimeRatesNonViolent = RiversidecrimeRatesNonViolent.drop_duplicates()
RiversidecrimeRatesNonViolentYears = RiversidecrimeRatesNonViolent.loc[RiversidecrimeRatesNonViolent["year"].isin(years)]

# drop extraneous columns
RiversidecrimeRatesNVComparison = RiversidecrimeRatesNonViolentYears.drop(columns=["key", "id", "place_name"])
# sort year in ascending order
RiversidecrimeRatesNVComparison = RiversidecrimeRatesNVComparison.sort_values(['year'], ascending=[True])

# Create seperate dataframes for each type of crime:
Riversidelarceny = RiversidecrimeRatesNVComparison.loc[RiversidecrimeRatesNVComparison["crime_type"] == "Larceny"]
# set the index of each dataframe to the year
Riversidelarceny = Riversidelarceny.set_index(["year"])

Riversidemotor = RiversidecrimeRatesNVComparison.loc[RiversidecrimeRatesNVComparison["crime_type"] == "Motor vehicle theft"]
Riversidemotor = Riversidemotor.set_index(["year"])

Riversideburglary = RiversidecrimeRatesNVComparison.loc[RiversidecrimeRatesNVComparison["crime_type"] == "Burglary"]
Riversideburglary = Riversideburglary.set_index(["year"])

Riversiderobbery = RiversidecrimeRatesNVComparison.loc[RiversidecrimeRatesNVComparison["crime_type"] == "Robbery"]
Riversiderobbery = Riversiderobbery.set_index(["year"])

RiversidepropCrime = RiversidecrimeRatesNVComparison.loc[RiversidecrimeRatesNVComparison["crime_type"] == "Property crime"]
RiversidepropCrime = RiversidepropCrime.set_index(["year"])

# Create a stacked bar plot where each crime type is in a distinct color:

# indices of each dataframe are the same size, can use arbitrary one for plot x-axis
Riverside_x_axis = Riversidelarceny.index
# each bar is a line here
p1 = plt.bar(Riverside_x_axis, Riversidelarceny["value"], color='purple', alpha=0.5, align="center")
p2 = plt.bar(Riverside_x_axis, Riversidemotor["value"].values, color='red', bottom=Riversidelarceny["value"].values)
p3 = plt.bar(Riverside_x_axis, Riversideburglary["value"].values, color='black', bottom=(Riversidemotor["value"] + Riversidelarceny["value"]).values)
p4 = plt.bar(Riverside_x_axis, Riversiderobbery["value"].values, color="#ADD8E6", bottom = (Riversidemotor["value"] + Riversidelarceny["value"] + Riversideburglary["value"]).values)
p5 = plt.bar(Riverside_x_axis, RiversidepropCrime["value"].values, color='pink', bottom = (Riversidemotor["value"] + Riversidelarceny["value"] + Riversideburglary["value"] + Riversiderobbery["value"]).values)

tick_locations = [value for value in Riverside_x_axis]
plt.xticks(tick_locations, years)
plt.title("Riverside, California Non-Violent Crime Rate: \n Comp Over Pre- and Post-Marijuana Legalization Years")
plt.ylabel("Crime Rate (per 100,000 People)")
plt.legend((p1[0], p2[0], p3[0], p4[0], p5[0]), ("Larceny", "Motor-vehicle Theft", "Burglary", "Robbery", "Property Crime"))
RiversidecrimeRatesNVComparison
```
